# Central Spin System — Levels 0–7

A structured overview of the progressive models of the central spin system in quantum dots, covering electron, nuclear, and interaction dynamics from basic Zeeman splitting to advanced decoupling control.

---

## Level 0 — Single Driven Electron (Baseline Control)

### System / Motivation
- Calibrate and understand coherent control of an isolated electron spin qubit.
- Benchmark for quantum control sequences (π, π/2 pulses, Rabi, Ramsey).

### Hamiltonian
$$
\hat{H} = \omega_e \hat{s}_z + \omega_{1,\mathrm{MW}} [ \cos(\omega_{\mathrm{MW}} t) \hat{s}_x + \sin(\omega_{\mathrm{MW}} t) \hat{s}_y ]
$$

### Typical Initial States
- Electron spin up: $|\uparrow\rangle_e$

### Key Control Concepts (Formal)
- **Rabi oscillations:** coherent transitions between $|\uparrow\rangle$ and $|\downarrow\rangle$.
- **Resonance condition:** $\omega_{\mathrm{MW}} = \omega_e$
- **Detuning:** $\Delta = \omega_e - \omega_{\mathrm{MW}}$

### Expected Dynamics / Results
- On resonance: full Rabi oscillations at frequency $\Omega_R = \omega_{1,\mathrm{MW}}$.
- Off-resonance: reduced oscillation amplitude and effective frequency $\Omega_{\mathrm{eff}} = \sqrt{\Delta^2 + \Omega_R^2}$.

### What It Means
- Establishes the basis for coherent spin manipulation in all higher-level experiments.

### Observables
- Electron population: $\langle \hat{s}_z(t) \rangle$
- Transverse coherence: $|\langle \hat{s}_+(t) \rangle|$

### Simulation Notes
- Use rotating-wave approximation (RWA).
- Extract $T_2^*$ from Ramsey sequences or decay of $\langle \hat{s}_+(t) \rangle$.

---

## Level 1 — Electron + Static Polarized Nuclei (Overhauser Shift)

### System / Motivation
- Include the static effect of nuclear polarization on the electron’s Zeeman energy.
- No nuclear dynamics yet.

### Hamiltonian
$$
\hat{H} \approx \omega_e \hat{s}_z + \sum_j \omega_N \hat{I}_{z,j} + \hat{s}_z B_{\mathrm{OH}}, \quad B_{\mathrm{OH}} = \sum_j a_j \langle \hat{I}_{z,j} \rangle
$$

### Typical Initial States
- Electron up, all nuclei up: $|\uparrow\rangle_e |\uparrow\uparrow\dots\uparrow\rangle_n$

### Key Control Concepts (Formal)
- Overhauser field $B_{\mathrm{OH}}$ shifts electron resonance frequency.
- Neglect transverse hyperfine terms $\hat{s}_x\hat{I}_{x,j}$ and $\hat{s}_y\hat{I}_{y,j}$.

### Expected Dynamics / Results
- Static shift of ESR resonance by $B_{\mathrm{OH}}$.
- No spin flips; pure dephasing if nuclear field fluctuates.

### What It Means
- The electron feels an effective magnetic field from the nuclei — the Overhauser field.

### Observables
- ESR resonance frequency shift.
- Electron precession phase accumulation.

### Simulation Notes
- Treat hyperfine interaction as static offset.
- Use ensemble averaging to model Overhauser noise.

---

## Level 2–3 — Homogeneous Central Spin System (Collective Mode Dynamics)

### System / Motivation
- Introduce dynamic hyperfine interaction between electron and N identical nuclei.
- Examine collective electron–nuclear coupling and coherent exchange.

### Hamiltonian
$$
\hat{H} = \omega_e \hat{s}_z + \omega_N \hat{I}_{\mathrm{tot},z} + a (\hat{s} \cdot \hat{I}_{\mathrm{tot}})
$$

where $\hat{I}_{\mathrm{tot}} = \sum_j \hat{I}_j$.

### Typical Initial States
- Electron up, nuclei fully polarized $|W_0\rangle = |\uparrow\uparrow\dots\uparrow\rangle$
- Single collective nuclear excitation $|W_1\rangle = \frac{1}{\sqrt{N}}\sum_j |\uparrow\dots\downarrow_j\dots\uparrow\rangle$

### Key Control Concepts (Formal)
- **Collective coupling:** $g_{\mathrm{col}} = (a/2)\sqrt{N}$
- **Bright mode:** symmetric nuclear excitation coupling to electron.
- **Dark modes:** orthogonal, non-coupled nuclear subspace.

### Expected Dynamics / Results
- Rabi-like oscillations between electron and collective nuclear mode.
- Avoided crossing of width $a\sqrt{N}$ at resonance $\omega_e \approx \omega_N$.

### What It Means
- Demonstrates the collective enhancement of coupling strength in a homogeneous system.

### Observables
- Electron spin: $\langle \hat{s}_z(t) \rangle$
- Nuclear collective polarization: $\langle \hat{I}_{\mathrm{tot},z}(t) \rangle$
- Bright-mode amplitude: $|b_0(t)|^2 = |\frac{1}{\sqrt{N}}\sum_j \langle \hat{I}_{+,j}\rangle|^2$

### Simulation Notes
- Restrict to symmetric subspace with dimension N+1.
- Exact diagonalization feasible for moderate N.

## Level 4 — Inhomogeneous Couplings (Realistic Quantum Dot)

### System / Motivation
- Introduce realistic nonuniform hyperfine couplings $a_j$.
- Study dephasing and partial loss of collectivity.

### Hamiltonian
$$
\hat{H} = \omega_e \hat{s}_z + \sum_j \omega_N \hat{I}_{z,j} + \sum_j a_j (\hat{s} \cdot \hat{I}_j)
$$

### Typical Initial States
- Electron up; nuclei mostly up with one or few flipped spins.

### Key Control Concepts (Formal)
- **Dephasing:** due to inhomogeneous Overhauser fields.
- **Dark modes:** become populated; partial decoherence of bright mode.

### Expected Dynamics / Results
- Free induction decay: $|\langle \hat{s}_+(t) \rangle| \sim e^{-(t/T_2^*)^2}$ with $T_2^* \sim 1/\sigma_B$.
- Partial transfer fidelity, non-reversible spin exchange.

### What It Means
- Real systems exhibit finite coherence even without external noise.

### Observables
- $\langle \hat{s}_+(t) \rangle$, site-resolved $\langle \hat{I}_{z,j}(t) \rangle$, spectral linewidth.

### Simulation Notes
- Exact for small N.
- Use mean-field or semiclassical bath models for large N.

---

## Level 5 — Driven Dynamics: Sidebands & Hartmann–Hahn

### System / Motivation
- Introduce time-dependent drives to engineer electron–nuclear resonance conditions.
- Enable coherent state transfer and spin-wave generation.

### Hamiltonian
$$
\hat{H} = \omega_e \hat{s}_z + \sum_j \omega_N \hat{I}_{z,j} + \sum_j a_j (\hat{s}\cdot\hat{I}_j)
+ \omega_{1,\mathrm{MW}}[\cos(\omega_{\mathrm{MW}}t)\hat{s}_x + \sin(\omega_{\mathrm{MW}}t)\hat{s}_y]
+ \sum_j \omega_{1,\mathrm{RF}}\cos(\omega_{\mathrm{RF}}t)\hat{I}_{x,j}
$$

### Typical Initial States
- Electron prepared in qubit superposition; nuclei polarized.

### Key Control Concepts (Formal)
- **Carrier resonance:** $\omega_{\mathrm{MW}} = \omega_e$
- **Sideband resonance:** $\omega_{\mathrm{MW}} = \omega_e \pm \omega_N$
- **Hartmann–Hahn condition:** $\Omega_R \approx \omega_{1,\mathrm{RF}}$

### Expected Dynamics / Results
- Carrier drive → electron Rabi oscillations.
- Sideband drive → electron–nuclear flip-flops (spin-wave excitation).
- Hartmann–Hahn → efficient polarization transfer.

### What It Means
- Enables coherent mapping between electron qubit and nuclear ensemble.

### Observables
- Electron $\langle \hat{s}_z(t) \rangle$, nuclear $\langle \hat{I}_z(t) \rangle$, ESR spectra with sidebands.

### Simulation Notes
- Use rotating frames and RWA.
- Time-dependent solvers for $\hat{H}(t)$.

## Level 6 — Nuclear–Nuclear Dipolar Interactions (Spin Diffusion)

### System / Motivation
- Include nuclear–nuclear dipolar couplings causing intrinsic decoherence.
- Quantify memory decay due to spin diffusion.

### Hamiltonian
$$
\hat{H} = \omega_e \hat{s}_z + \sum_j \omega_N \hat{I}_{z,j} + \sum_j a_j (\hat{s}\cdot\hat{I}_j)
+ \sum_{j<k} b_{jk}[\hat{I}_j\cdot\hat{I}_k - 3(\hat{I}_j\cdot\hat{r}_{jk})(\hat{I}_k\cdot\hat{r}_{jk})]
$$

Secular approximation:
$$
\hat{H}_{dip} \approx \sum_{j<k} b_{jk}[\hat{I}_{z,j}\hat{I}_{z,k} - \tfrac{1}{4}(\hat{I}_{+,j}\hat{I}_{-,k}+\hat{I}_{-,j}\hat{I}_{+,k})]
$$

### Typical Initial States
- Collective spin-wave $|W_1\rangle$
- Partially polarized random nuclear bath

### Key Control Concepts (Formal)
- **Spin diffusion:** flip-flop processes redistribute excitation.
- **Dipolar dephasing:** destroys collective phase coherence.

### Expected Dynamics / Results
- Decay of bright-mode overlap: $F(t) = |\langle W_1|e^{-iH_{dip}t}|W_1\rangle|^2$
- Redistribution of nuclear polarization over time.

### What It Means
- Sets the fundamental coherence limit of nuclear quantum memory.

### Observables
- Bright-mode amplitude $|b_0(t)|^2$, local $\langle I_{z,j}(t)\rangle$, correlations $\langle I_{+,j}I_{-,k}\rangle$

### Simulation Notes
- Exact simulation for small N only.
- Use cluster-correlation or semiclassical methods for larger ensembles.

---

## Level 7 — Dynamical Decoupling and Coherence Protection

### System / Motivation
- Introduce pulse-based or continuous control to protect coherence.
- Extend memory lifetime beyond intrinsic dipolar limits.

### Hamiltonian
$$
\hat{H}(t) = \hat{H}_{static} + \hat{H}_{ctrl}(t)
$$
where
$$
\hat{H}_{static} = \omega_e\hat{s}_z + \sum_j \omega_N\hat{I}_{z,j} + \sum_j a_j(\hat{s}\cdot\hat{I}_j) + \hat{H}_{dip}
$$
and $\hat{H}_{ctrl}(t)$ describes control pulses or continuous driving fields.

### Typical Initial States
- Electron superposition $|+\rangle = (|\uparrow\rangle+|\downarrow\rangle)/\sqrt{2}$
- Stored nuclear bright-mode excitation $|W_1\rangle$

### Key Control Concepts (Formal)
- **Refocusing:** π pulses reverse phase accumulation.
- **Average Hamiltonian theory:** cancel unwanted interactions.
- **Dynamical decoupling:** suppress low-frequency noise.
- **CHASE protocol:** composite nuclear pulses refocusing hyperfine and dipolar terms.

### Expected Dynamics / Results
- Coherence lifetime extended to $T_2^{DD} \gg T_2$.
- Slow decay of bright-mode amplitude |b₀(t)|².
- Enhanced readout fidelity of stored states.

### What It Means
- Dynamical control converts the spin ensemble into a long-lived quantum memory.
- Enables practical quantum interfaces with high-fidelity state transfer and retrieval.

### Observables
- Electron coherence $|\langle \hat{s}_+(t) \rangle|$
- Memory fidelity $F_{readout}$
- Nuclear bright-mode survival $|b_0(t)|^2$

### Simulation Notes
- Model pulses as instantaneous unitaries or short Hamiltonians.
- Use Floquet or Magnus expansion for average Hamiltonian analysis.
- Extract extended $T_2^{DD}$ by fitting coherence decay.
